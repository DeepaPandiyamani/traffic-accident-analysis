Task 5:Analyze traffic accident data to identify patterns related to road conditions, weather, and time of day. Visualize accident hotspots and contributing factors.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/traffic.csv'
data = pd.read_csv(file_path)


print(data.info())
print(data.head())



plt.figure(figsize=(12, 6))
sns.countplot(x='Vehicle_driver_relation', data=data, palette='viridis')
plt.title('Number of Accidents by Vehicle Driver Relation')
plt.xlabel('Vehicle Driver Relation')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(12, 6))
sns.countplot(x='Driving_experience', data=data, palette='magma')
plt.title('Number of Accidents by Driving Experience')
plt.xlabel('Driving Experience')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(x='Types_of_Junction', data=data, palette='plasma')
plt.title('Number of Accidents by Types of Junction')
plt.xlabel('Types of Junction')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(x='Weather_conditions', data=data, palette='coolwarm')
plt.title('Number of Accidents by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(x='Cause_of_accident', data=data, palette='cubehelix')
plt.title('Number of Accidents by Cause of Accident')
plt.xlabel('Cause of Accident')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.show()

categorical_columns = ['Vehicle_driver_relation', 'Driving_experience', 'Types_of_Junction', 'Weather_conditions', 'Cause_of_accident']
factorized_data = data[categorical_columns].apply(lambda x: pd.factorize(x)[0])

corr_matrix = factorized_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
